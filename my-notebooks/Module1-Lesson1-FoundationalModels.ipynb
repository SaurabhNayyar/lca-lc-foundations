#### Langchain's create_agent

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage

agent=create_agent(model="gpt-5-nano")

response = agent.invoke({"messages":[HumanMessage(content="What is the capital of Moon?")]})
```

```python
pprint(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke({"messages":[HumanMessage(content="What is the capital of Moon?"),
                                     AIMessage(content="The capital of Moon is Lunar City."),
                                   HumanMessage(content="What is the population of Lunar City?")]})    
```

##### Streaming

```python
for token,metadata in agent.stream({"messages":[HumanMessage(content="The capital of moon is Lunar city,what is the population of Lunar City?")]},stream_mode="messages"):
    if token.content:
        print(token.content, end='', flush=True)
```

##### Streaming Example - Streaming Mode="updates"

```python
from langchain.agents import create_agent


def get_weather(city: str) -> str:
    """Get weather for a given city."""

    return f"It's always sunny in {city}!"

agent = create_agent(
    model="gpt-5-nano",
    tools=[get_weather],
)
for chunk in agent.stream(  
    {"messages": [{"role": "user", "content": "What is the weather in SF?"}]},
    stream_mode="updates",
):
    print(f"chunk: {chunk}")
    for step, data in chunk.items():
        print(f"step: {step}")
        print(f"content: {data['messages'][-1].content_blocks}")
        
```

##### Streaming Mode = "messages"

```python
from langchain.agents import create_agent


def get_weather(city: str) -> str:
    """Get weather for a given city."""

    return f"It's always sunny in {city}!"

agent = create_agent(
    model="gpt-5-nano",
    tools=[get_weather],
)
for token, metadata in agent.stream(  
    {"messages": [{"role": "user", "content": "What is the weather in SF?"}]},
    stream_mode="messages",
):
    # print(f"node: {metadata['langgraph_node']}")
    # print(f"content: {token.content_blocks}")
    # print("\n")
    print(f"metadata: {metadata}")
    print(f"token: {token}")
    print("\n")
```

```python
system_prompt = """You are fictional science writer, expert in given fictional capital names of the places.
                    User: What is the capital of Jupiter?
                    Agent: Jupitology

                    User: What is the capital of Earth?
                    Agent: Earthopolis
"""

agent = create_agent(model="gpt-5-nano",
                     system_prompt=system_prompt)

human_Message = HumanMessage(content="What is the capital of Moon?")

response = agent.invoke({"messages":[human_Message]})

pprint(response)
```

```python
from pydantic import BaseModel

class Capital(BaseModel):
    name: str
    vibe: str
    location: str
    economy: str

system_prompt = "You are a fictional scifi writer."
human_Message = HumanMessage(content="What is the capital of Moon?")

agent = create_agent(model = "gpt-5-nano",
                     system_prompt=system_prompt,
                     response_format=Capital)

response = agent.invoke({"messages":[human_Message]})

pprint(response)
```

```python
capital_info = response['structured_response']

print(f"Name : {capital_info.name} \n Vibe:{capital_info.vibe} \n Location:{capital_info.location} \n Economy:{capital_info.economy}")
```
