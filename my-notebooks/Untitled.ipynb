```python
from pprint import pprint
```

#### Langchain's init_chat_model

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-5-nano")

response = model.invoke("What is the capital of Moon?")
```

```python
pprint(response)
```

#### Langchain's create_agent

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage

agent=create_agent(model="gpt-5-nano")

response = agent.invoke({"messages":[HumanMessage(content="What is the capital of Moon?")]})
```

```python
pprint(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke({"messages":[HumanMessage(content="What is the capital of Moon?"),
                                     AIMessage(content="The capital of Moon is Lunar City."),
                                   HumanMessage(content="What is the population of Lunar City?")]})    
```

```python
pprint(response['messages'][-1].content)
```

##### Streaming

```python
for token,metadata in agent.stream({"messages":[HumanMessage(content="The capital of moon is Lunar city,what is the population of Lunar City?")]},stream_mode="messages"):
    if token.content:
        print(token.content, end='', flush=True)
```
